```python
import cicada.math
import numpy
```

```python
f = cicada.math.Field(modulus=251)
f
```

```python
f.modulus, f.fieldbits, f.fieldbytes
```

```python
generator = numpy.random.default_rng(seed=1234)
a = f.uniform(size=(2,2), generator=generator)
a
```

```python
b = f.uniform(size=(2, 2), generator=generator)
b
```

```python
f.subtract(a, b)
```

```python
c = cicada.math.FieldArray([[0, 1], [2, 3]], modulus=251)
c
```

```python
f.subtract(a, c)
```
