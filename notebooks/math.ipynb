```python
import numpy

from cicada.math import Field, FieldArray
from cicada.encoder import FixedFieldArray
```

```python
f = Field(modulus=251)
f
```

```python
f.modulus, f.fieldbits, f.fieldbytes
```

```python
generator = numpy.random.default_rng(seed=1234)
a = f.uniform(size=2, generator=generator)
a
```

```python
b = f.uniform(size=2, generator=generator)
b
```

```python
f.subtract(a, b)
```

```python
c = FieldArray([0, 1], modulus=251)
c
```

```python
f.subtract(a, c)
```
