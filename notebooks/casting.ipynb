```python
import logging
logging.basicConfig(level=logging.DEBUG)

import numpy

from cicada.math import Field, FieldArray
from cicada.encoder import BinaryFieldArray, FixedFieldArray, BinaryFieldEncoder, FixedFieldEncoder
```

```python
# Create a FieldArray.
a = FieldArray([1, 2, 3], modulus=251)
a, a.modulus
```

```python
# Cast a FieldArray to BinaryFieldArray.
b = a.view(BinaryFieldArray)
b, b.modulus
```

```python
# Copy a FieldArray to BinaryFieldArray.
c = BinaryFieldArray(a)
c, c.modulus
```

```python
# Cast a FieldArray to FixedFieldArray.
d = a.view(FixedFieldArray)
d, d.modulus, d.precision
```

```python
# Copy a FieldArray to FixedFieldArray.
e = FixedFieldArray(a, precision=4)
e, e.modulus, e.precision
```

```python
# Modify the original FieldArray.
a[0] = 4
a, a.modulus
```

```python
# The cast should also change.
b, b.modulus
```

```python
# The copy should remain the same.
c, c.modulus
```

```python
# The cast should also change.
d, d.modulus, d.precision
```

```python
# The copy should remain the same.
e, e.modulus, e.precision
```
