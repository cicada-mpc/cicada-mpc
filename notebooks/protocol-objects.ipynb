```python
import logging

import numpy

from cicada import additive
from cicada.arithmetic import Field
from cicada.communicator import SocketCommunicator
from cicada.encoding import FixedPoint
from cicada.logging import Logger
from cicada.przs import PRZSProtocol

logging.basicConfig(level=logging.DEBUG)
```

```python
def main(communicator):
    logger = Logger(logging.getLogger(), communicator)
    field = Field(order=127)
    encoding = FixedPoint(precision=2)
    przs = PRZSProtocol(communicator=communicator, field=field, seed=communicator.rank)
    share = additive.share(przs=przs, communicator=communicator, src=0, secret=numpy.array(3), shape=(), field=field, encoding=encoding)
    secret = additive.reveal(communicator=communicator, share=share, dst=None, field=field, encoding=encoding)
    logger.info(f"Player {communicator.rank} share: {share}")
    logger.info(f"Player {communicator.rank} secret: {secret}")

SocketCommunicator.run(fn=main, world_size=3);
```
